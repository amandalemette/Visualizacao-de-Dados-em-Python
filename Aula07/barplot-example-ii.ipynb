```python
x = np.arange(5)
x
```

```python
poupanca = [1500, 32000, 54000, 2300, 91000]
poupanca
```

```python
plt.bar(x,poupanca,color = 'limegreen',edgecolor = 'teal',linewidth = 2,\
       hatch = '*')
plt.ylabel('Poupana (R$)',fontsize = 18)
plt.savefig('GraficoBarras.png',dpi = 600,bbox_inches = 'tight')
plt.show()
```

```python
pessoas = ['Ana','Pedro','Paulo','Carla', 'Andre']
pessoas
```

```python
plt.bar(x,poupanca,color = 'limegreen',edgecolor = 'teal',linewidth = 2,\
       hatch = '*')
plt.xticks(x,pessoas,rotation = 45)
plt.ylabel('Poupana (R$)',fontsize = 18)
plt.savefig('GraficoBarras.png',dpi = 600,bbox_inches = 'tight')
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
def mil(x,pos):
    return f'R$ {x*1e-3:.1f}k'
```

```python
formatter = FuncFormatter(mil)
```

```python
fig,ax = plt.subplots()
ax.yaxis.set_major_formatter(formatter)
barras = plt.bar(x,poupanca,color = 'mediumorchid',edgecolor = 'plum',linewidth = 2,\
       hatch = '*')
barras[0].set_hatch('+')
barras[1].set_hatch('o')
barras[4].set_hatch('/')
plt.xticks(x,pessoas,rotation = 45)
plt.ylabel('Poupana',fontsize = 18)
plt.savefig('GraficoBarras.png',dpi = 600,bbox_inches = 'tight')
plt.show()
```
